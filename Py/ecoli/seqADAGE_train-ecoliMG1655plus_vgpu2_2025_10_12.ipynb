```python


comps = ["ecoli_MG165plus_genome_part2_log_counts_norm_01"]





model_dict_post = {}

stime = time.time()
ltime = 0
c = 0



seeds = list(range(10))
seeds = random.sample(range(10000), 10)

for comp in comps:
    for seed in seeds:
        name = 'ad_' + comp + '_2025_09_01_' + str(seed+571)
        print(name)
        ttime = time.time()
        mseq = run_model.run_model('../data_files/ecoli_MG165plus_genome_part2_log_counts_norm_01.csv',
                                                  seed=seed+571,
                                                  enc_dim = 450,
                                                  kl1=0,
                                                  kl2=0,
                                                  act = "tanh",
                                                  act2="tanh",
                                                  tied = True,
                                                  epochs=100,
                                                  init="glorot_uniform",
                                                  batch_size=15,
                                                  dropout = 0,
                                                  mm = 0.6,
                                                  lr = 0.5)
       
        model_dict_post[name] = mseq
        ltime = ((time.time() - ttime) + ltime)
        c+=1

rtime = time.time() - stime
```
